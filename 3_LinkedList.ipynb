```python
#Linked Lists
class Node:
    def __init__(self, data=None, next=None, prev = None):
        self.data = data
        self.next = next
        self.prev = prev

class LinkedList:
    def __init__(self):
        self.head = None
        self.prev = None
    def insert_at_begining(self, data):
        node = Node(data, self.head, None)
        if self.head != None:
            self.head.prev = node
        else:
            self.prev = node
        self.head = node
    def print(self):
        if self.head is None:
            print("Linked List is empty")
            return
        itr = self.head
        llstr = ''
        while itr:
            llstr += str(itr.data) + '-->'
            itr = itr.next
        print(llstr)
    def insert_at_end(self, data):
        node = Node(data, None, self.prev)
        if self.prev != None:
            self.prev.next = node
        else:
            self.head = node
        self.prev = node
    def insert_values(self,data_list):
        self.head = None
        for data in data_list:
            self.insert_at_end(data)
            #print(data)
    def get_length(self):
        count = 0
        itr = self.head
        while itr:
            count+=1
            itr = itr.next
        return count
    def remove_at(self, index):
        selfLen = self.get_length()
        #print(index, selfLen)
        if index <0 or index >= selfLen:
            raise Exception("Invalid index")
        if index == 0:
            self.head = self.head.next
            self.head.prev = None
            return
        if index == selfLen-1:
            self.prev = self.prev.prev
            self.prev.next = None
            return
        count = 0
        itr = self.head
        while itr:
            if count == index:
                itr.next.prev = itr.prev
                itr.prev.next = itr.next
                return
            itr = itr.next
            count += 1
    def insert_at(self, index, data):          
        selfLen = self.get_length()
        if index <0 or index >= selfLen:
            raise Exception("Invalid index")
        if index == 0:
            self.insert_at_begining(data)
            return
        if index == selfLen:
            self.insert_at_end(data)
            return
        itr = self.head
        lastItr = itr
        count = 1
        while itr:
            itr = itr.next
            if count == index:
                #print(itr.data, lastItr.data)
                node = Node(data, itr, lastItr)
                lastItr.next = node
                itr.prev = node
                return
            lastItr = itr
            count += 1
    def insert_after_value(self, data_after,data_to_insert):            
        selfLen = self.get_length()
        nextSelf = self.head.next
        if self.head.data == data_after:
            node = Node(data_to_insert, nextSelf, self.head)
            nextSelf.prev = node
            self.head.next = node
            return
        itr = self.head
        while itr:
            itr = itr.next
            if itr.data == data_after:
                #print(itr.data)
                if itr == self.prev:
                    self.insert_at_end(data_to_insert)
                    return
                node = Node(data_to_insert, itr.next, itr)
                itr.next.prev = node
                itr.next = node
                return
        #search for first occuance of data-after value in ll
        #insert data_to_insert after data_after node
    
    def remove_by_value(self, data):                                      
        # Remove first node that contains data
        if (self.head.data == data):
            self.head = self.head.next
            self.head.prev = None
            return
        itr = self.head
        lastItr = itr
        while itr:
            itr = itr.next
            if itr.data == data:
                if itr.next == None:
                    self.prev = itr.prev
                    self.prev.next = None
                    return
                itr.next.prev = lastItr
                lastItr.next = itr.next
                return
            lastItr = itr
    def print_forward(self):                                 
        self.print()
    def print_backward(self):                                
        if self.head is None:
            print("Linked List is empty")
            return
        itr = self.prev
        llstr = ''
        while itr:
            llstr += str(itr.data) + '-->'
            itr = itr.prev
        print(llstr)
            
            
        
if __name__ == '__main__':
    ll = LinkedList()
    ll.insert_values(["banana","mango","grapes","orange"])
    #ll.print()
    #ll.insert_at_begining("kiwi")
    ll.print_backward()
```
